Countries with largest number of players

```python
nationality = df20['nationality'].value_counts()
nationality
```

```python
plt.figure(figsize=(5,3))
plt.bar(list(nationality.keys()[0:5]) , list(nationality[0:5]) , color = ['orange' ,'green' ,'yellow', 'blue', 'red'])
plt.show()
```

Salary of the Players

```python
player_salary = df20[['short_name' , 'wage_eur']]
player_salary
```

```python
player_salary=player_salary.sort_values (by=['wage_eur' ], ascending=False)
player_salary.head()
```

```python
plt.figure(figsize=(8,4)) 
plt.bar(list (player_salary['short_name'])[0 : 5],list (player_salary['wage_eur'])[0 : 5], color=['r','g','y','b','c']) 
plt.show() 
# fifa['nationality'] == 'Germany'
```
